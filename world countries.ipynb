```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
url = 'https://www.scrapethissite.com/pages/simple/'
```

```python
r = requests.get(url)
r
```

```python
soup = BeautifulSoup(r.text, 'lxml')
```

```python
titles = soup.find_all('h3', class_ = 'country-name')
countries = []
for i in titles:
    country = i.text
    countries.append(country)
countries
```

```python
capitals = soup.find_all('span', class_ = 'country-capital')
capital_list = []
for i in capitals:
    capital = i.text
    capital_list.append(capital)
capital_list
```

```python
populations = soup.find_all('span', class_ = 'country-population')
pop_list = []
for i in populations:
    pop = i.text
    pop_list.append(pop)
pop_list
```

```python
areas = soup.find_all('span', class_ = 'country-area')
area_list = []
for i in areas:
    area = i.text
    area_list.append(area)
area_list
```

```python
df = pd.DataFrame({'Country': countries, 'Capital': capital_list, 'Population':pop_list, 'Area (km2)': area_list})
df.head()
```

```python
df['Country'] = df['Country'].str.strip('\n\n, \n')
df
```

```python
df.to_csv('World Countries.csv', index = False)
```
